```
# üß† Import library
from youtube_transcript_api import YouTubeTranscriptApi

# üîó ‡∏•‡∏¥‡∏á‡∏Å‡πå YouTube ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏î‡∏∂‡∏á transcript
video_url = "https://www.youtube.com/watch?v=GKCBpj9FQXU"  # ‡πÄ‡∏õ‡∏•‡∏µ‡πà‡∏¢‡∏ô‡∏•‡∏¥‡∏á‡∏Å‡πå‡πÑ‡∏î‡πâ
video_id = video_url.split("v=")[-1]

# üéØ ‡∏î‡∏∂‡∏á transcript
try:
    transcript = YouTubeTranscriptApi.get_transcript(video_id)

    # ‚úÖ ‡πÅ‡∏™‡∏î‡∏á transcript ‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÄ‡∏ß‡∏•‡∏≤
    print("‚úÖ Transcript:\n")
    for entry in transcript:
        print(f"{entry['start']:.2f}s - {entry['text']}")

    # üíæ ‡∏£‡∏ß‡∏°‡∏Ç‡πâ‡∏≠‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
    full_text = '\n'.join([entry['text'] for entry in transcript])

    # üíæ ‡πÄ‡∏ã‡∏ü‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏ü‡∏•‡πå .txt
    with open("transcript.txt", "w", encoding="utf-8") as f:
        f.write(full_text)
    print("\nüíæ Transcript saved as 'transcript.txt'")

except Exception as e:
    print("‚ùå Error:", e)
```
